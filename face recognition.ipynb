```python
def faces_analysis_pics(p1, p2):
    img1_representation = model.predict(preprocess_image(p1))[0,:]
    img2_representation = model.predict(preprocess_image(p2))[0,:]
    img1_representation = l2_normalize(img1_representation)
    img2_representation = l2_normalize(img2_representation)
    print("Cosine distance: ",findCosineDistance(img1_representation, img2_representation))
    if findCosineDistance(img1_representation, img2_representation) < 0.10:
        print("These are same people")
    else:
        print("These are different people")
    print("Euclidean distance: ",findEuclideanDistance(img1_representation, img2_representation))
    if findEuclideanDistance(img1_representation, img2_representation) < 0.50:
        print("These are same people")
    else:
        print("These are different people")	
    displayImages(p1, p2)
```
